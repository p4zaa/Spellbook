```python
from pathlib import Path
original_gif_directory = "/Users/pa/Documents/Pa's side projects/Science of Going Viral/icons"
destination_directory = "/Users/pa/Documents/Pa's side projects/Science of Going Viral/icons/removed_background"

Path(destination_directory).mkdir(parents=True, exist_ok=True)

original_gif_paths = sb_utils.get_file_paths(
    path=original_gif_directory,
    file_extension=".gif"
)

destination_gif_paths = sb_utils.get_file_paths(
    path=destination_directory,
    file_extension=".gif"
)

destination_gif_path_names = [path.name for path in destination_gif_paths]

'''# Drop duplicate GIFs
for original_path in original_gif_paths:
    if original_path.name in destination_gif_path_names:
        destination_gif_paths.remove(original_path)
        print(f"Removed duplicate GIF: {original_path.name}")'''

# Remove white background from GIFs
for gif_path in original_gif_paths:
    if gif_path.name not in destination_gif_path_names:
        sb_utils.image_processing.remove_white_background(
            input_path=gif_path,
            output_path=f'{destination_directory}/{gif_path.stem}',
            threshold=240,
            exact_white=False,
            output_format="GIF"
        )
        print(f"Processed GIF: {gif_path.name}")
    else:
        print(f"Skipped existing GIF: {gif_path.name}")
```
